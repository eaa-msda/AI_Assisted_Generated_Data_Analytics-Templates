```python
# Impport Necessary Lirbaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset (LOCATION OF DATASET)
df = pd.read_csv('/content/UniversalBank data.csv')
```

```python
# Basic Exploration
df.head()           # Display first 5 rows
df.tail()           # Display last 5 rows
df.info()           # Overview of dataset
df.describe()       # Summary statistics
df.columns          # List of column names
df.shape            # Dimensions of dataset
df.isnull().sum()   # Check for missing values
df.dtypes           # Data types of each column
```

```python
# Handling Missing Data
df.dropna(inplace=True)           # Remove missing values
df.fillna(0, inplace=True)    # Fill missing values with 0, for example
```

```python
# Data Cleaning (Un-Hastag, Put relevant variables AS NEEDED)
# df.rename(columns={'old_name':'new_name'}, inplace=True)  # Rename columns
# df.drop(['column_name'], axis=1, inplace=True)            # Drop columns
df.duplicated().sum()                                     # Check for duplicates
df.drop_duplicates(inplace=True)                          # Remove duplicates
```

```python
# PAY ATTENTTION TO WHAT YOU NEED!
```

```python
# Univariate Analysis
# Chosen Variables:
# Age – To see the age distribution of customers.
# Income – To observe the income levels.
# CreditCard – To check how many have a credit card.

# Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()

# Income Distribution
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Income'])
plt.title('Income Distribution')
plt.show()

# Credit Card Ownership Distribution
sns.countplot(x='CreditCard', data=df)
plt.title('Credit Card Ownership Distribution')
plt.show()
```

```python
# Bivariate Analysis
# Chosen Variables:
# Age vs Personal Loan – To explore how age influences personal loan acquisition.
# Income vs Personal Loan – To see if higher income impacts loan approval.
# Education vs Personal Loan – Educational level's effect on loan decisions.

import matplotlib.pyplot as plt
import seaborn as sns

# Age vs Personal Loan
plt.figure(figsize=(8,5))
sns.boxplot(x='Personal Loan', y='Age', data=df)
plt.title('Age vs Personal Loan')
plt.show()

# Income vs Personal Loan
plt.figure(figsize=(8,5))
sns.boxplot(x='Personal Loan', y='Income', data=df)
plt.title('Income vs Personal Loan')
plt.show()

# Education vs Personal Loan
plt.figure(figsize=(8,5))
sns.countplot(x='Education', hue='Personal Loan', data=df)
plt.title('Education vs Personal Loan')
plt.show()
# Education vs Personal Loan – Educational level's effect on loan decisions.

# Age vs Personal Loan
plt.figure(figsize=(8,5))
sns.boxplot(x='Personal Loan', y='Age', data=df)
plt.title('Age vs Personal Loan')
plt.show()

# Income vs Personal Loan
plt.figure(figsize=(8,5))
sns.boxplot(x='Personal Loan', y='Income', data=df)
plt.title('Income vs Personal Loan')
plt.show()

# Education vs Personal Loan
plt.figure(figsize=(8,5))
sns.countplot(x='Education', hue='Personal Loan', data=df)
plt.title('Education vs Personal Loan')
plt.show()
```

```python
# Categorical Data Visualization
# Chosen Variables:
# Education – To see how education level varies among customers.
# Securities Account – To analyze how many have securities accounts.
# CD Account – To observe certificate of deposit ownership.

# Education Distribution
sns.countplot(x='Education', data=df)
plt.title('Education Distribution')
plt.show()

# Securities Account Ownership
sns.countplot(x='Securities Account', data=df)
plt.title('Securities Account Ownership')
plt.show()

# CD Account Ownership
sns.countplot(x='CD Account', data=df)
plt.title('CD Account Ownership')
plt.show()
```
